```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on ='Country Code', how='left')
final_df.head()
```

```python
# To check data types

final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# Pie chart -- Top 3 countries

plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

observation : Maximum records or transactions for zomato is from India. 
    After that USA and then UK.

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

Observation : 
      1. Rating between 4.5 to 4.9 --> Excellent
      2. Rating between 4.0 to 4.4 --> Very good
      3. Rating between 3.5 to 3.9 --> Good
      4. Rating between 2.5 to 3.4 --> Average
      5. Rating between 1.8 to 2.4 --> Poor

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data= ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data= ratings,palette=['blue','red','orange','yellow','green','green'])
```

observation :
    1. Not rated vount is very high
    2. Maximum Number of rating is between 2.5 to 3.4

```python
# count plot

sns.countplot(x="Rating color",data = ratings,palette=['blue','red','orange','yellow','green'])
```
